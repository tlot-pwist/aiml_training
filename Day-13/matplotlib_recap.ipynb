```python
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','Mar','Apr']
sales=[2000,3000,2500,3600]
```

```python
#plt.plot(x,y,color='color_name',linestyle=line_style',linewidth=value,marker='marker_symbol',labrl='label_name')
plt.plot(months,sales,color='blue',linestyle='--',linewidth=2,marker='o',label='Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.title('Month Sales Report')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')
plt.xlabel('X Lines')
plt.ylabel('Y Lines')
plt.title('Simple Lines Plot')
plt.legend(loc='upper left')
plt.show()
```

```python
categories=['Electronics','Mobiles','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['red','blue','green','purple','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Categories wise Sales Report')
plt.grid(axis='y',linestyle='--',color='green',alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
categories=['Electronics','Mobiles','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['red','blue','green','purple','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='red',label='Sales Info')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Categories wise Sales Report')
plt.grid(axis='y',linestyle='--',color='green',alpha=0.5)
plt.xticks(rotation=30, fontsize=12, color='magenta')
plt.yticks([20000,40000,60000,80000,100000],['20K','40K','60K','80K','1M'],fontsize=12, color='green')
plt.legend(title='Legend',fontsize=8)
plt.tight_layout()
plt.show()
```

```python
height_cm=[150,144,155,160,170,185]
weight_kg=[50,45,60,48,70,90]

plt.figure(figsize=(8,5))
plt.scatter(height_cm,weight_kg,color='red',edgecolor= None,label='height vs weight')
plt.xlabel('Height(cm)')
plt.ylabel('weight(kg)')
plt.title('height vs weight')
plt.grid(axis='y',linestyle='--',color='green',alpha=0.5)
plt.xticks([150,144,155,160,170,185],['150cm','144cm','155cm','160cm','170cm','185cm'],rotation=30, fontsize=12, color='magenta')
plt.yticks([50,45,60,48,70,90],['50kg','45kg','60kg','48kg','70kg','90kg'],fontsize=12, color='green')
plt.legend(title='Legend',fontsize=8)
plt.tight_layout()
plt.show()
```

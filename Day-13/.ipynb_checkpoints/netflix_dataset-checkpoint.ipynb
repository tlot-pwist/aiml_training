```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
```

```python
df.head()
```

```python
df=df.dropna(subset=['release_year'])
df['release_year']=df['release_year'].astype(int)
plt.figure(figsize=(8,5))
plt.hist(
    df['release_year'],
    bins=100,color='green',
    edgecolor='black',
    alpha=0.8
    )
plt.title('netflix release titles by year')
plt.xlabel('release year')
plt.ylabel('number of titles')
plt.grid(
    color='blue',
    linestyle='--',
    linewidth=0.7
)
plt.tight_layout()
plt.savefig(
    'netflix_bar.pdf',
    dpi=300,
    bbox_inches='tight'
)
plt.show()
```

```python
df=df.dropna(subset=['release_year'])
df['release_year']=df['release_year'].astype(int)
movies=df[df['type']=='Movie']
shows=df[df['type']=='TV Show']
plt.figure(figsize=(8,5))
plt.hist(
    df['release_year'],
    bins=100,color='green',
    edgecolor='black',
    alpha=0.8,
    label=['Movies','TV Shows']
    )
plt.title('netflix release titles by year')
plt.xlabel('release year')
plt.ylabel('number of titles')
plt.grid(
    color='blue',
    linestyle='--',
    linewidth=0.7
)
plt.tight_layout()
plt.legend(title='Movies/TV Shows')
plt.savefig(
    'netflix_showsvsmovies.pdf',
    dpi=300,
    bbox_inches='tight'
)
plt.show()
```

```python
df=df.dropna(subset=['type'])
type_counts=df['type'].value_counts()
print(type_counts)
movies=df[df['type']=='Movies']
shows=df[df['type']=='TV Shows']
print(f'number of movies in this netflix dataset: {len(movies)}')
print(f'number of movies in this netflix dataset: {len(shows)}')
```

```python
plt.figure(figsize=(7,7))
plt.pie(
    type_counts,
    labels=type_counts.index,
    colors=['blue','yellow'],
    autopct='%1.2f%%',
    startangle=140, 
    explode=(0.05,0)
)
plt.title('Movies and TV Shows in Netflix')
plt.axis('equal')
plt.legend(title='')

plt.savefig(
    'netflix_showsmovies_pie.pdf',
    dpi=300,
    bbox_inches='tight'
)
plt.show()
```

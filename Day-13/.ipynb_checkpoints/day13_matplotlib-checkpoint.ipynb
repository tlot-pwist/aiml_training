```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
temp=np.random.randint(20,100,size=50)
plt.scatter(x,y,c=temp,cmap='plasma',s=120,edgecolor='black')
plt.colorbar(label='temperature')
plt.title('scatter plot colored by temperature')
plt.xlabel('x-random')
plt.ylabel('y-random')
plt.show()
```

```python
height_cm=[150,144,155,160,170,185]
weight_kg=[50,45,60,48,70,90]
age=[18,21,33,51,28,23]
```

```python
plt.figure(figsize=(8,5))
```

```python
plt.scatter(height_cm,weight_kg,c=age,cmap='rainbow',s=120,edgecolor='black')
plt.colorbar(label='age(years)')
plt.title('scatter plot: height(cm) vs weight(kg) (colored by age)')
plt.xlabel('height(cm)')
plt.ylabel('weight(kg)')
plt.grid(color='blue',linestyle='--',linewidth=0.7)
plt.show()
```

```python
#plt.pie(x,labels='',colors='',autoopct='',startagnle='',shadow='',explode='',radius='')
departments=['HR','Web_dev','App-Dev','Agile-scrum','Testing']
expenses=[12000,25000,18000,15000,20000]
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(expenses,labels=departments,colors=plt.cm.rainbow([0.1,0.3,0.5,0.7,0.9]),autopct='%1.1f%%',startangle=140,explode=(0,0.05,0,0,0))
plt.title('Departments wise Expense Report')
plt.show()
```

##### create pie plot using the following data
##### sales_data={'Mobile':45000, 'Clothing':30000, 'Groceries':35000, 'Home Decor':20000}
##### show % with two decimal points
##### use colormap to give different color to you pie

```python
sales_data={'Mobile':45000, 'Clothing':30000, 'Groceries':35000, 'Home Decor':20000}
plt.figure(figsize=(7,7))
```

```python
explode=(0,0.05,0,0,)
colors=plt.cm.rainbow([0.1,0.3,0.5,0.7,0.9]),

labels=sales_data.keys()
sizes=sales_data.values()
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.2f%%',
    startangle=140, 
    explode=explode
)
plt.title('Sales Data')
plt.axis('equal')
plt.show()
```

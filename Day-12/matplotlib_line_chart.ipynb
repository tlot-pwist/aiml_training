# create a line chart from our data

##### use file students.xlsx for fetching the data
##### generate line chart based on student name and student marks

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_excel(r'C:\Users\DELL\Coding\aiml_training\Day-10\students.xlsx')
```

```python
names=df['Name']
marks=df['Marks']
```

```python
df
```

```python
plt.plot(names,marks,marker='s',color='green')
plt.title('student marks line chart')
plt.xlabel('student name')
plt.ylabel('student marks')
plt.grid(True)
plt.show()
```

```python
#dash lines style
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='o',color='orange',linestyle='--')
plt.title('student progress chart')
plt.xlabel('-- Name --')
plt.ylabel('-- Marks --')
#plt.grid(True)
plt.show()
```

```python
#dotted lines styyle
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='o',color='blue',linestyle=':')
plt.title('student progress chart')
plt.xlabel('-- Name --')
plt.ylabel('-- Marks --')
#plt.grid(True)
plt.show()
```

```python
df_patients['arrival_date']=pd.to_datetime(df_patients['arrival_date'])
#print(df_patients['arrival_date'])
avg_age=df_patients.groupby('arrival_date')['age'].mean().sort_values(ascending=False)

#age_counts=df_patients['age'].value_counts().sort_index()
plt.figure(figsize=(100,25))
#bar_colors=['blue','red','green','yellow','orange','pink']
plt.bar(avg_age.index,avg_age.values,color='orange',width=0.8,edgecolor=None)
plt.title('Number of age by arrival date')
plt.xlabel('arrival_date')
plt.ylabel('Age')
plt.show()
```

```python
# create a bar chart using below data
# create a vertical bas graph where each bar represents the value of pythonLib respective liked by users
# every bar must be different color
pythonLib=['Numpy','Pandas','Matplotlib','Scikit-learn']
popular=[5000,7000,9000,4000]
plt.figure(figsize=(6,4))
bar_colors=['blue','red','green','yellow']
plt.bar(pythonLib,popular,color=bar_colors,width=0.5,edgecolor=None)
plt.title('Popularity of Python Library')
plt.xlabel('Python Library')
plt.ylabel('Popularity')
plt.show()
```
